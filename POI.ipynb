```python
def map_values(value):
    if value in Short:
        return 'ST'
    elif value in Long:
        return 'LT'
    elif value in Cool:
        return 'CF'
    else:
        return None  # If the value in column 'A' is not in any of the lists

# Add a new column 'New_Column' using the apply method
poi_class['Category'] = poi_class['groupname'].apply(map_values)
```

```python
# Function to convert the string representation of coordinates to a Point object
def get_geometry(row):
    coordinates_str = row['geometry'].replace('POINT (', '').replace(')', '')
    x, y = map(float, coordinates_str.split())
    return Point(x, y)

# Apply the function to create the new 'geometry' column
df['geometry'] = df.apply(get_geometry, axis=1)

# # Convert the DataFrame to a GeoDataFrame
# # gdf = gpd.GeoDataFrame(poi_class, geometry='geometry')



# # # Function to convert the string representation of coordinates to longitude and latitude columns
# # def get_longitude_latitude(row):
# #     coordinates_str = row['geometry'].replace('POINT (', '').replace(')', '')
# #     longitude, latitude = map(float, coordinates_str.split())
# #     return pd.Series({'Longitude': longitude, 'Latitude': latitude})


# # poi_class[['Longitude', 'Latitude']] =poi_class.apply(get_longitude_latitude, axis=1)

# def get_longitude_latitude(row):
#     coordinates_str = row['geometry'].replace('POINT (', '').replace(')', '')
#     longitude, latitude = map(float, coordinates_str.split())
#     return pd.Series({'Longitude': longitude, 'Latitude': latitude})

# poi_class[['Longitude', 'Latitude']] = poi_class.apply(get_longitude_latitude, axis=1)
```
